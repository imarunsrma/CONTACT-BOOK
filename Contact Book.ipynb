```python
import tkinter as tk
from tkinter import messagebox

# Contact Book class
class ContactBook:
    def _init_(self):
        self.contacts = {}  # Dictionary to store contacts
    
    def add_contact(self, name, phone_number):
        self.contacts[name] = phone_number
        return f"Contact '{name}' added successfully."

    def view_contacts(self):
        if not self.contacts:
            return "No contacts found."
        return "\n".join([f"Name: {name}, Phone: {phone}" for name, phone in self.contacts.items()])

    def search_contact(self, name):
        return self.contacts.get(name, "Contact not found.")
    
    def update_contact(self, name, new_phone):
        if name in self.contacts:
            self.contacts[name] = new_phone
            return f"Contact '{name}' updated successfully."
        return "Contact not found."
    
    def delete_contact(self, name):
        if name in self.contacts:
            del self.contacts[name]
            return f"Contact '{name}' deleted successfully."
        return "Contact not found."

# Functions for each button action
def add_contact():
    name = entry_name.get()
    phone = entry_phone.get()
    if name and phone:
        message = contact_book.add_contact(name, phone)
        messagebox.showinfo("Add Contact", message)
    else:
        messagebox.showerror("Error", "Please enter both name and phone number.")
    entry_name.delete(0, tk.END)
    entry_phone.delete(0, tk.END)

def view_contacts():
    contacts = contact_book.view_contacts()
    messagebox.showinfo("Contact List", contacts)

def search_contact():
    name = entry_name.get()
    if name:
        contact = contact_book.search_contact(name)
        messagebox.showinfo("Search Contact", f"Name: {name}, Phone: {contact}")
    else:
        messagebox.showerror("Error", "Please enter a name to search.")
    entry_name.delete(0, tk.END)

def update_contact():
    name = entry_name.get()
    new_phone = entry_phone.get()
    if name and new_phone:
        message = contact_book.update_contact(name, new_phone)
        messagebox.showinfo("Update Contact", message)
    else:
        messagebox.showerror("Error", "Please enter both name and new phone number.")
    entry_name.delete(0, tk.END)
    entry_phone.delete(0, tk.END)

def delete_contact():
    name = entry_name.get()
    if name:
        message = contact_book.delete_contact(name)
        messagebox.showinfo("Delete Contact", message)
    else:
        messagebox.showerror("Error", "Please enter a name to delete.")
    entry_name.delete(0, tk.END)

# Initialize ContactBook instance
contact_book = ContactBook()

# Set up the main application window
app = tk.Tk()
app.title("Contact Book")
app.geometry("300x400")

# Labels and Entries for Contact Information
label_name = tk.Label(app, text="Name:")
label_name.pack()
entry_name = tk.Entry(app)
entry_name.pack()

label_phone = tk.Label(app, text="Phone Number:")
label_phone.pack()
entry_phone = tk.Entry(app)
entry_phone.pack()

# Buttons for Contact Book Operations
btn_add = tk.Button(app, text="Add Contact", command=add_contact)
btn_add.pack(pady=5)

btn_view = tk.Button(app, text="View Contacts", command=view_contacts)
btn_view.pack(pady=5)

btn_search = tk.Button(app, text="Search Contact", command=search_contact)
btn_search.pack(pady=5)

btn_update = tk.Button(app, text="Update Contact", command=update_contact)
btn_update.pack(pady=5)

btn_delete = tk.Button(app, text="Delete Contact", command=delete_contact)
btn_delete.pack(pady=5)

# Run the application
app.mainloop()
```
